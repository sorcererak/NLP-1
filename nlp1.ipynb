```python
import numpy as np
import matplotlib.pyplot as plt
def norm(point1,point2):
    return np.linalg.norm(np.array(point2)-np.array(point1))
def check(point1,point2,d,r):
    i=0
    while i<d:
        if abs(point1[i]-point2[i])<r:
            i=i+1
        else:
            return 0
    return 1
d=int(input("Enter any dimension:"))
points=np.random.uniform(0,100,size=(1000,d))
pairs=[]
r=int(input("Enter any number between 1 and 10:"))
i=0
while i<1000:
    j=i+1
    while j<1000:
        if check(points[i],points[j],d,r):
            if norm(points[i],points[j])<r:
                pairs.append([points[i],points[j]])
        j=j+1
    i=i+1
print(pairs)
if d==2:
    x_points=points[:,0]
    y_points=points[:,1]
    colours=np.random.randint(0,100,(1000,1))
    plt.scatter(x_points,y_points,c=colours,cmap='viridis')
    plt.grid()
    plt.show()
```
